```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def smooth_function(x):
    return (1 / (math.sqrt(2 * math.pi))) * math.exp(-1/2 * (x  ** 2))

def non_smooth_function(x):
    return math.exp(-abs(x))
```

```python
def smooth_L_n(data, x):
    result = 0
    for k in range(len(data)):
        p_r = smooth_function(data[k])
        for j in range(len(data)):
            if j != k:
                p_r *= (x - data[j]) / (data[k] - data[j])
        result += p_r
    return result

def non_smooth_L_n(data, x):
    result = 0
    for k in range(len(data)):
        p_r = non_smooth_function(data[k])
        for j in range(len(data)):
            if j != k:
                p_r *= (x - data[j]) / (data[k] - data[j])
        result += p_r
    return result
```

```python
data_1 = [x for x in range(-1, 2)]
data_2 = [x for x in range(-5, 6)]
data_3 = [x for x in range(-10, 11)]
```

```python
def make_paint(data, name, original_func, interpolate_func, size = (10, 6)):
    x = np.linspace(min(data), max(data), 100)
    y = [interpolate_func(data, x_i) for x_i in x]
    plt.figure(figsize=size)
    plt.plot(x, y, label='Interpolation Polynomial')
    plt.plot(x, [original_func(x_i) for x_i in x], label=name)
    plt.scatter(data, [original_func(x) for x in data], color='red', label='Interpolation Points')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Lagrange Interpolation Polynomial')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python

make_paint(data_1, 'Normal Distribution', smooth_function, smooth_L_n)
```

```python

make_paint(data_2, 'Normal Distribution', smooth_function, smooth_L_n)
```

```python

make_paint(data_3, 'Normal Distribution', smooth_function, smooth_L_n)
```

```python

make_paint(data_1, 'Non Smooth Function', non_smooth_function, non_smooth_L_n)
```

```python

make_paint(data_2, 'Non Smooth Function', non_smooth_function, non_smooth_L_n)
```

```python
make_paint(data_3, 'Non Smooth Function', non_smooth_function, non_smooth_L_n)
```
